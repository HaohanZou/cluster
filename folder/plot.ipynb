```python
max_length = loss_2.shape[0]

def pad_1(obj, target):
    if len(obj) < target:
        add_temp = np.ones(target - len(obj))
        add_temp = pd.Series(add_temp)
        obj=pd.concat([obj, add_temp])
    return obj


loss_1_b = pd.DataFrame(pad_1(loss_1_b, max_length), columns = ['best'])
loss_1_q = pd.DataFrame(pad_1(loss_1_q, max_length), columns = ['quantile'])


loss_3_b = pd.DataFrame(pad_1(loss_3_b, max_length), columns = ['best'])
loss_3_q = pd.DataFrame(pad_1(loss_3_q, max_length), columns = ['quantile'])

loss_4_b = pd.DataFrame(pad_1(loss_4_b, max_length), columns = ['best'])
loss_4_q = pd.DataFrame(pad_1(loss_4_q, max_length), columns = ['quantile'])

plt.figure(figsize=(10, 6))
agents = np.array(loss_2.index)
plt.plot(agents, loss_1_b, label = "trail 1")

plt.plot(agents, loss_2_b, label='trail 2')

plt.plot(agents, loss_3_b, label='trail 3')

plt.plot(agents, loss_4_b, label='trail 4')


# Add labels and title

plt.xlabel('epoch')
plt.ylabel('Reward')
plt.title('Best Reward of each Trail')
plt.legend()
plt.grid(True)
plt.ylim((0.75,1.1))
plt.show()
```
